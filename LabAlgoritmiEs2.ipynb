# Cenni Teorici
## Algoritmi di ordinamento
Il processo di ordinamento rappresenta una questione cruciale nell'ambito dell'informatica e dell'analisi degli algoritmi. Si tratta dell'operazione di riorganizzazione di un gruppo di elementi in una sequenza ben disposta secondo un particolare criterio, che può implicare un ordinamento crescente o decrescente. L'obiettivo primario dell'ordinamento consiste nell'ottenere una disposizione dei dati che sia agevole per le operazioni di ricerca, confronto e analisi.

L'importanza del problema di ordinamento è evidente in vari contesti applicativi, tra cui l'elaborazione di database, la gestione di enormi quantità di dati, la risoluzione di problemi di ottimizzazione, e altre situazioni simili. La presenza di un algoritmo di ordinamento efficiente può contribuire in modo sostanziale al miglioramento delle performance di altre operazioni e algoritmi che dipendono dall'organizzazione corretta degli elementi.

Esistono numerosi algoritmi di ordinamento, ciascuno dei quali adotta strategie e approcci diversi per organizzare gli elementi. Questi algoritmi possono variare notevolmente in termini di complessità computazionale, velocità di esecuzione e adattabilità a tipi specifici di dati o scenari particolari. La selezione dell'algoritmo di ordinamento più idoneo dipende dal contesto specifico e dai requisiti particolari dell'applicazione.


## Insertion sort
L'Insertion Sort è un algoritmo di ordinamento elementare che si basa sul concetto dell'inserimento graduale di elementi in una lista. È un metodo intuitivo ed è particolarmente adatto per ordinare piccoli array o situazioni in cui l'array è già parzialmente ordinato.

Il funzionamento dell'Insertion Sort si basa su un approccio incrementale. L'algoritmo valuta uno alla volta gli elementi dell'array e, in ogni passo, inserisce l'elemento corrente nella sua corretta posizione all'interno della porzione già ordinata dell'array. Inizia considerando il primo elemento dell'array come una sequenza ordinata di un solo elemento. Successivamente, procede con l'elemento successivo e lo colloca nella sua posizione appropriata all'interno della sequenza già ordinata, spostando gli elementi più grandi verso destra.

Per comprenderne meglio il funzionamento, si può immaginare di ordinare una mano di carte in ordine crescente. Si inizia con una carta nella mano (corrispondente al primo elemento dell'array) e poi si prende una nuova carta (rappresentante il secondo elemento dell'array) e la si posiziona correttamente nella mano, spostando le carte più grandi verso destra. Questo processo si ripete per ogni carta successiva fino a quando tutte le carte sono state sistemate nella loro giusta posizione.

In ogni istante, le carte nella mano sinistra sono ordinate; inizialmente erano le prime della pila di carte sul tavolo. Dal punto di vista del codice, questo significa che durante l'i-esima iterazione, i primi $i-1$ elementi sono già ordinati tra loro.
Si noti che l'algoritmo ordina  **sul posto**, vale a dire che i numeri sono risistemati ordinatamente all'interno dello stesso array di input.

Vediamo adesso il codice python che implementa l'algoritmo insertion sort.

```python
# Codice di insertion sort
def insertionSort(A):
    for j in range(1, len(A)):
        key = A[j]
        i = j - 1
        while i >= 0 and A[i]>key:
            A[i + 1] = A[i]
            i = i - 1
        A[i + 1] = key
```

## Complessità 
* **Caso migliore** (array già ordinato): in una situazione ottimale in cui l'array è già completamente ordinato, l'Insertion Sort richiede solamente una singola scansione dell'array senza necessità di effettuare spostamenti. In questo scenario, l'algoritmo dimostra una complessità temporale di $O(n)$.
* **Caso peggiore** (array ordinato inversamente): in situazioni in cui l'array è disposto in ordine inverso, l'algoritmo di Insertion Sort richiede di spostare ogni elemento una posizione verso destra per ogni elemento precedente. Questa operazione implica una complessità temporale quadraticamente crescente, espressa con la notazione $O(n^2)$.
* **Caso medio**: nel contesto del caso medio, l'andamento dell'Insertion Sort è più efficiente rispetto al caso peggiore, ma non raggiunge la complessità ideale del caso migliore. La complessità media di Insertion Sort rimane $O(n^2)$.

Ecco che quindi, insertion sort risulta una soluzione valida per quantità di dati limitate, ma al contempo risulta inefficiente per ordinare grandi quantità di dati.

Vediamo adesso una tabella che riassume le varie casistiche:
| Caso migliore | Caso medio | Caso peggiore |
|-----------------------|-----------------------|-----------------------|
| $O(n)$             | $O(n^2)$               | $O(n^2)$              |

## Correttezza
L'Insertion sort è affidabile, nel senso che assicura che l'array di input sarà in una sequenza ordinata dopo l'esecuzione dell'algoritmo. Il concetto fondamentale dell'Insertion Sort è di mantenere una sottosequenza di elementi ordinati mentre si scorre attraverso l'array, inserendo ciascun elemento nella sua posizione appropriata all'interno di questa sottosequenza. Questo processo di inserimento ripetuto garantisce che l'intero array sarà alla fine disposto in modo corretto. Visto che l'algoritmo è implementato in forma iterativa, è possibile dimostrare la sua affidabilità mediante un'invariante di ciclo.
## Stabilità
L'Insertion Sort è un algoritmo di ordinamento che mantiene la stabilità, il che implica che preserva l'ordine relativo degli elementi con chiavi identiche. In altre parole, se due elementi hanno la stessa chiave e sono originariamente disposti in posizioni diverse all'interno dell'array di input, l'applicazione dell'Insertion Sort garantirà che questi due elementi mantengano il medesimo ordine relativo nella sequenza ordinata risultante.


## Counting sort
Counting sort è un algoritmo che non si basa direttamente sulla comparazione diretta tra valori, ma piuttosto -come per altro suggerisce il nome- 
sul conteggio di essi.
E' un algoritmo che non ordina sul posto, e che riesce ad ordinare in un tempo lineare, anche se per ottenere questo necessita di memoria aggiuntiva.
Per funzionare, counting sort suppone che ciascuno degli $n$ elementi in input sia un numero intero compreso nell'intervallo da $0$ a $k$, per qualche intero $k$. Ecco che quando $k=O(n)$, allora l'ordinamento viene effettuato con complessità $O(n)$.

Il procedimento del Counting Sort coinvolge tre fasi interconnesse. Per prima cosa, si crea un array di conteggio con una dimensione equivalente all'intervallo massimo dei valori presenti nell'input. Questo array funge da registro delle occorrenze, e durante l'analisi dell'input, il contatore associato a ciascun valore viene incrementato. Alla fine di questa fase, l'array di conteggio conterrà il numero di occorrenze per ciascun valore presente nell'input.
Successivamente, si utilizza l'array di conteggio appena popolato. Esso viene utilizzato per determinare le posizioni finali di ciascun valore nell'output ordinato. Questo si realizza sommando i valori precedenti nell'array di conteggio, in modo che l'elemento $i-esimo$ nell'array ausiliario di posizioni rappresenti la posizione finale dell'ultimo elemento con valore $i$ nell'output.
Infine, nell'ultima fase si riattraversa l'input e si genera l'array di output ordinato. Per ciascun elemento nell'array di input, si sfrutta l'array ausiliario di posizioni per identificare la sua posizione finale nell'output ordinato. L'elemento viene quindi inserito in questa posizione, e al tempo stesso, il valore corrispondente nell'array di conteggio viene decrementato. Questo passo è essenziale per mantenere un'adeguata traccia delle occorrenze rimanenti durante la costruzione dell'output finale e ordinato.

Nonostante le sue qualità, il counting sort presenta alcune restrizioni. La sua efficacia è condizionata da un intervallo limitato dei valori di input e può comportare l'utilizzo di una notevole quantità di memoria, specialmente quando l'intervallo è ampio. Di conseguenza, risulta particolarmente vantaggioso in contesti specifici in cui queste limitazioni non costituiscono un ostacolo, ad esempio, nell'ordinamento di numeri interi non negativi all'interno di un intervallo ristretto, mentre non è proprio applicabile quando l'intervallo dei valori sia abbastanza ampio.

<figure>
  <img src="https://favtutor.com/resources/images/uploads/mceu_87381749051625725474967.png" alt="Esempio di applicazione di counting  sort"  style="width: 70%; height:auto">
  <figcaption>Figura 2: esempio di applicazione di counting sort</figcaption>
</figure>


```python
# Codice di counting sort
def countingSort(A):
    # Crea l'array ordinato B
    B = [0] * len(A)
    # Trova il valore massimo nell'array di input così da avere la dimensione per l'array di conteggio
    maxValue = max(A)
    
    # Crea un array di conteggio con dimensioni pari al valore massimo + 1, per tenere conto di tutti i valori
    C = [0] * (maxValue + 1)   

    # Conta il numero di occorrenze di ciascun elemento nell'array
    for num in A:
        C[num] += 1

    # Conta il numero di elementi minori o uguali a i
    # Questo array indica la posizione in cui ogni valore deve essere posizionato nell'output finale
    for i in range(1, maxValue + 1):
        C[i] += C[i - 1]

    for num in A:
        B[C[num] - 1] = num
        C[num] -= 1
    return B
```

## Complessità
La performance del counting sort è influenzata sia dalla dimensione dell'array da ordinare, cioè $n$, sia dalla gamma di valori possibili nell'array, rappresentata da $k$ (dove $k$ è il massimo valore dei valori presenti).

La complessità temporale dell'algoritmo è lineare, e cioè $O(n+k)$.

Nel caso peggiore, la complessità temporale è lineare rispetto alla somma di $n$ e $k$. Ciò significa che il tempo e lo spazio necessari per eseguire l'algoritmo crescono linearmente all'aumentare di entrambi i fattori. La complessità peggiore si verifica quando i valori nell'array sono distribuiti uniformemente su un ampio intervallo. Ad ogni modo, la disposizione dei valori nell'array di input non influisce sulla prestazione dell'algoritmo, come era per insertion sort, e dunque non c'è nessuna netta distinzione di casistiche temporali in questo caso. In altre parole, a prescindere dall'input, la complessità rimane lineare. Va sottolineato ad ogni modo che se $k$ è notevolmente inferiore a $n$," l'algoritmo può dimostrarsi molto efficiente. Tuttavia, quando $k$ è paragonabile o persino superiore a $n$, l'efficienza dell'algoritmo può declinare.
In ultimo, occorre mettere in chiaro che per poter confrontare il funzionamento dei due algoritmi, si è optato per un'analisi di counting sort negli stessi casi migliore-medio-peggiore definiti per insertion sort, anche se come già detto per counting sort in ognuno di essi la complessità rimane uguale.

## Correttezza
Il Counting Sort è un algoritmo affidabile e assicura con certezza che l'array di input sarà ordinato in modo accurato. Questo è possibile grazie al fatto che il Counting Sort fa uso delle informazioni sull'ordine dei valori presenti, senza la necessità di confronti diretti tra gli elementi. Questo approccio garantisce l'accuratezza del processo di ordinamento. Essendo un algoritmo di forma iterativa, è possibile dimostrarne la correttezza tramite un'invariante di ciclo.

## Stabilità
Il Counting Sort gode della proprietà di stabilità: quando ci sono elementi con chiavi identiche, esso assegna loro posizioni nell'array ordinato basandosi sull'ordine di apparizione nell'array di input. In questo modo, assicura che due elementi con la stessa chiave mantengano sempre il loro ordine relativo nella sequenza ordinata.


# Descrizione dei test
Vogliamo adesso confrontare il funzionamento due algoritmi. Analizzeremo ciascuno nei 3 casi di riferimento, e cioè:
* **caso migliore**: genereremo un array già ordinato e lo daremo in input ad entrambi gli algritmi
* **caso medio**: genereremo un array con numeri casuali e lo daremo in input ad entrambi gli algoritmi
* **caso peggiore**: genereremo un array ordinato inversamente e lo daremo in input ad entrambi gli algoritmi
  
Tutti i test sono effettuati nella funzione *test()*, e per ottenere risultati più attendibili e precisi sono eseguiti un numero di volte *numTests=100*,   e poi mediati.
Gli array su cui lavoreranno i due algoritmi avranno una dimensione via via crescente, partendo da 1 ed arrivando a *dimMax=2000*: in questa maniera si osserverà il comportamento di insertion sort e counting sort in relazione alla dimensione dell'input.

## Librerie utilizzate 

```python
import numpy as np #usata per generare array
import random #usata per generare numeri casuali
from timeit import default_timer as timer #usata per misurare i tempi di esecuzione
import matplotlib.pyplot as plt #usata per generare grafici
import matplotlib.patches as mpatch #usata per aggiungere delle legende ai grafici
```

## Esecuzione dei test
Specifichiamo che:
* **numTests** (val. 100) è il numero di volte che si eseguono i test;
* **dimMax** (val. 2000) è la dimensione massima degli array su cui operano gli algoritmi;
* **step** (val. 100) è il passo di incremento della dimensione degli array ad ogni iterazione.
Per misurare i tempi di esecuzione inoltre, si è fatto affidamento alla libreria *timeit*, ed in particolare alla sua funzione *timer()*: la si è quindi chiamata sia prima che dopo ogni operazione da misurare, e si è fatto la sottrazione dei due valori. Il risultato è stato sempre moltiplicato per $1000$ in modo tale da esprimerlo in [ms].

È fondamentale notare che l'algoritmo counting sort viene eseguito per primo sull'array, seguito dall'insertion sort. Questa sequenza è dovuta al fatto che il counting sort restituisce un nuovo array con gli elementi ordinati, preservando l'array originale, mentre l'insertion sort effettua l'ordinamento direttamente sullo stesso array di partenza (sul posto). In caso contrario, se invertissimo l'ordine delle esecuzioni, l'insertion sort verrebbe applicato all'array già ordinato generato dal counting sort, cosa che falserebbe i risultati dell'esperimento.
Di seguito vediamo il codice della funzione principale di test:

```python
def test(numTests, dimMax , step):
    resultsInsertionSortBestCase  = np.zeros(int(dimMax/step))
    resultsInsertionSortWorstCase = np.zeros(int(dimMax/step))
    resultsInsertionSortAverageCase = np.zeros(int(dimMax/step))
    resultsCountingSortBestCase = np.zeros(int(dimMax/step))
    resultsCountingSortWorstCase = np.zeros(int(dimMax/step))
    resultsCountingSortAverageCase = np.zeros(int(dimMax/step))

    for i in range(0, numTests):
        k = 0
        for j in range (1, dimMax, step):
            bestCaseInput = [random.randint(1, j) for _ in range(j)]
            bestCaseInput = sorted(bestCaseInput)
            worstCaseInput = [random.randint(1, j) for _ in range(j)]
            worstCaseInput = sorted(worstCaseInput, reverse=True)
            averageCaseInput = np.random.randint(1, j+1, j)

            start = timer()
            countingSort(bestCaseInput)
            end = timer()
            resultsCountingSortBestCase[k] += ((end - start) * 1000)

            start = timer()
            insertionSort(bestCaseInput)
            end = timer()
            resultsInsertionSortBestCase[k] += ((end - start) * 1000)

            start = timer()
            countingSort(averageCaseInput)
            end = timer()
            resultsCountingSortAverageCase[k] += ((end - start) * 1000)

            start = timer()
            insertionSort(averageCaseInput)
            end = timer()
            resultsInsertionSortAverageCase[k] += ((end - start) * 1000)

            start = timer()
            countingSort(worstCaseInput)
            end = timer()
            resultsCountingSortWorstCase[k] += ((end - start) * 1000)

            start = timer()
            insertionSort(worstCaseInput)
            end = timer()
            resultsInsertionSortWorstCase[k] += ((end - start) * 1000)

            k = k + 1
    for i in range(0, resultsCountingSortWorstCase.size):
        resultsInsertionSortBestCase[i] /= numTests
        resultsInsertionSortWorstCase[i] /= numTests
        resultsInsertionSortAverageCase[i] /= numTests
        resultsCountingSortBestCase[i] /= numTests
        resultsCountingSortWorstCase[i] /= numTests
        resultsCountingSortAverageCase[i] /= numTests

    results = [resultsInsertionSortBestCase, resultsCountingSortBestCase,
               resultsInsertionSortAverageCase,
               resultsCountingSortAverageCase, resultsInsertionSortWorstCase,
               resultsCountingSortWorstCase]


    return results
```

A questo punto andiamo ad eseguire la funzione appena definita, impostando anche i parametri di test già descritti:

```python
numTests = 100
dimMax = 2000
step = 100
results = test(numTests, dimMax, step)
```

## Generazione delle tabelle dei tempi di esecuzione
Di seguito si riporta il codice che genera le tabelle dei tempi di esecuzione dei due algoritmi nei tre casi considerati, al crescere della quantità dei valori da ordinare.

```python
# Prende in input una lista di colonne di dati, una tupla di intestazioni e un titolo
def traceTables(columns: list, headers: tuple, title: str):
    fig, ax = plt.subplots(figsize=(8, 10))
    plt.title(title)

    # Unisci le liste di dati come colonne al fine di creare un array bidimensionale di dati: 'data'
    data = np.stack(tuple(columns), axis=1)

    # Stile tabella
    ax.axis('off')
    table = ax.table(cellText=data, colLabels=headers, loc='center', cellLoc='center')
    table.auto_set_column_width(col=list(range(len(columns))))
    table.scale(1, 1.5)

    # Colorazione headers e le righe pari
    cell_colors = {
        cell: ("#ffd1d1", {"weight": "bold"})
        if table[cell].get_text().get_text() in headers
        else ("#ffe4e4", {})
        for cell in table._cells
        if cell[0] % 2 == 0
    }
    for cell, (color, text_props) in cell_colors.items():
        # set the color of the cell
        table[cell].set_facecolor(color)
        # set the text properties of the cell
        table[cell].set_text_props(**text_props)
        # ** used to expand the dictionary
    plt.show()

def createTables(results,  dimMax, step):
    traceTables(
        *[
            [
                [i for i in range(1, dimMax, step)],
                ["{:.3e}".format(val) for val in results[0]],
                ["{:.3e}".format(val) for val in results[1]],
            ],
            (
                "Nr elementi",
                "Insertion sort",
                "Counting sort",
            ),
            "Figura 3: Caso migliore - tempi di esecuzione in [ms] con array già ordinato",
        ]
    )

    traceTables(
        *[
            [
                [i for i in range(1, dimMax, step)],
                ["{:.3e}".format(val) for val in results[2]],
                ["{:.3e}".format(val) for val in results[3]],
            ],
            (
                "Nr elementi",
                "Insertion sort",
                "Counting sort",
            ),
            "Figura 4: Caso medio - tempi di esecuzione in [ms] con array con numeri casuali",
        ]
    )

    traceTables(
        *[
            [
                [i for i in range(1, dimMax, step)],
                ["{:.3e}".format(val) for val in results[4]],
                ["{:.3e}".format(val) for val in results[5]],
            ],
            (
                "Nr elementi",
                "Insertion sort",
                "Counting sort",
            ),
            "Figura 5: Caso peggiore - tempi di esecuzione in [ms] con array ordinato inversamente",
        ]
    )
```

A questo punto andiamo ad eseguire la funzione appena definita, passandogli anche alcuni dei parametri di test indispensabili per generare correttamente la struttura delle tabelle:

```python
createTables(results, dimMax, step)
```

# Generazione dei grafici

# Osservazioni finali
## Caso migliore (array già ordinato):
Nel caso migliore dell'insertion sort, quando i dati sono già ordinati correttamente, si osserva una complessità lineare, come mostrato nel grafico a sinistra nella figura 6. Questo conferma le ipotesi teoriche iniziali, che indicavano una complessità nel caso migliore di $O(n)$.  Allo stesso tempo, il counting sort mantiene una complessità lineare, in linea con le previsioni (visibile in figura 7).

Inoltre, analizzando sia il grafico in figura 12 che la tabella nella figura 3, si nota che i tempi di esecuzione dell'insertion sort sono inferiori rispetto a quelli del counting sort, nonostante entrambi abbiano un comportamento lineare. Questa disparità è dovuta al fatto che l'insertion sort è un algoritmo più semplice del counting sort. Quest'ultimo richiede cicli esterni multipli, eseguendo così più operazioni a costo lineare, mentre l'insertion sort si limita a un singolo ciclo esterno, rendendolo più efficiente in termini di tempo.
## Caso medio (array con valori casuali):
Nel caso medio, cioè quello con array contenente valori casuali, l'andamento delle prestazioni dell'insertion sort, mostra una tendenza quadratica, come si può vedere nel grafico in figura 8. D'altro canto, il counting sort mantiene un comportamento lineare, come si può vedere nel grafico in figura 9.  All'aumentare del volume dei dati, il counting sort dimostra una superiorità significativa rispetto all'insertion sort in termini di prestazioni. Tuttavia, è importante notare che quando la quantità di dati da ordinare è limitata, l'insertion sort può, come già discusso in precedenza, presentare prestazioni migliori rispetto al counting sort.
## Caso peggiore (array ordinato inversamente):
Nel caso peggiore, vale a dire quando i dati sono disposti in ordine inverso, l'insertion sort ha un andamento quadratico, come evidenziato nel grafico nella Figura 10. Si nota inoltre che i tempi di esecuzione sono ancora più sfavorevoli rispetto al caso medio, quando si tratta di ordinare la stessa quantità di valori, come dimostra la differenza di picco nell'asse dei tempi nei grafici delle Figure 8 e 10, nonché dalle tabelle presentate nelle Figure 4 e 5. D'altra parte, il counting sort mantiene una complessità lineare e mostra tempi di esecuzione simili al caso migliore.

## Ulteriori osservazioni e conclusioni
I test eseguiti confermano le ipotesi iniziali di funzionamento dei due algoritmi.
Dunque la complessità dell'insertion sort è lineare nel caso migliore ed è quadratica nel caso medio e nel caso peggiore.
Al contrario, si è visto invece -come da previsione- che il counting sort non è influenzato dalla disposizione degli elementi nell'array da ordinare, e quindi in tutti e tre i casi mantiene una complessità lineare. Concludiamo ricordando che un risultato così efficiente è ottenibile dal counting sort grazie al fatto che utilizza dello spazio aggiuntivo (cosa che l'insertion sort non fa) e proprio per questo non è utilizzabile in tutti i contesti, ma il tutto dipende dal range di valori (e quindi dal valore massimo presente nell'array da ordinare). Se esso è troppo grande rispetto alla dimensione dell'array, l'algoritmo perde di efficienza, oppure può non essere addirittura applicabile per il troppo spazio occupato.
